```python
import torch
import torch.nn as nn
```

```python
# 查看nn.Module的__call__方法实现
print("nn.Module.__call__()的核心逻辑:")
print("-" * 50)
print("def __call__(self, *input, **kwargs):")
print("    # 1. 执行前向预处理钩子")
print("    # 2. 调用self.forward(*input, **kwargs)")
print("    # 3. 执行后向处理钩子")
print("    # 4. 返回结果")

# 实际验证
class DebugModule(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        print("forward()方法被调用!")
        return x * 2

    def __call__(self, *args, **kwargs):
        print("__call__()方法开始执行")
        print(f"参数: args={args}, kwargs={kwargs}")
        result = super().__call__(*args, **kwargs)
        print("__call__()方法结束执行")
        return result

debug_model = DebugModule()
print("\n实际调用过程:")
output = debug_model(torch.tensor([1, 2, 3]))
print(f"输出: {output}")
```
